# Graphs based analysis

```python
# Plot all numerical features

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
plt.suptitle("Univariate analysis of numerical features", fontsize=20, fontweight="bold", alpha = 0.8, y =1.)

for i in range(0, len(numerical_columns)):
    plt.subplot(2,3,i+1)
    sns.kdeplot(df, x=numerical_columns[i], fill = True, color = 'r')
    plt.xlabel(numerical_columns[i])
    plt.tight_layout()
```

```python
# Plot all categorical features

plt.figure(figsize=(15,15))
plt.suptitle("Univariate analysis of categorical features",fontsize=20, fontweight = "bold", alpha = 0.8, y=1.)

for i in range(0, len(categorical_columns)):
    plt.subplot(4,3,i+1)
    sns.countplot(df, y = categorical_columns[i], palette = "Set2")
    plt.ylabel(categorical_columns[i])
    #plt.xticks(rotation = 90)
    plt.tight_layout()

```

```python
# Bivariate analysis

# Education and Income count

df.groupby(['education'])["income"].value_counts().reset_index()
```

```python
# Bivariate analysis between workclass and income categories

work_inc_df = df.groupby(['workclass'])["income"].value_counts().reset_index()

plt.figure(figsize=(10,6))
plt.suptitle("Bivariate Analysis between education and income categories", fontsize=20, fontweight="bold", alpha=0.8, y=1.)
sns.barplot(work_inc_df, y="workclass",x="count", hue="income", palette="hls")
plt.tight_layout()
```

# Data Cleaning


```python
df.head()
```

```python
# We know that workclass, occupation and native-country has unknown or incorrect category "?"

# Replace "?" with most frequent category
```

```python
import statistics as stat

stat.mode(df["workclass"])
```

```python
df_copy = df
```

```python
df_copy.head()
```

```python
df_copy["workclass"] = df_copy["workclass"].str.replace(" ?", stat.mode(df_copy["workclass"]))
df_copy["occupation"] = df_copy["occupation"].str.replace(" ?", stat.mode(df_copy["occupation"]))
df_copy["native-country"] = df_copy["native-country"].str.replace(" ?", stat.mode(df_copy["native-country"]))
```
